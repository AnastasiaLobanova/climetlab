# NOAA's hurricane database

```python
import climetlab as  cml
```

```python
hurdat2 = cml.load_dataset("hurricane-database")
```

```python
hurdat2.home_page
```

Get the Pandas frame

```python
df = hurdat2.to_pandas()
```

## Huricanes making landfall at category 5

```python
df[(df.category==5) & (df.type=='L')]
```

## Get the track for Katrina

```python
katrina = df[(df.name=='katrina') & (df.year==2005)]
```

```python
katrina
```

## Plot the track

```python
cml.plot_map(katrina, style='cyclone-track', margins=1.5)
```

## Retrieve matching fields from ERA5

```python
source = cml.load_source("cds", 
                         "reanalysis-era5-single-levels",
                         variable="msl",
                         product_type= "reanalysis",
                         based_on={"data": katrina, "extent": "bounding-box", "margins": 1.5})

for s in source:
    cml.plot_map(s, width=400)
```
