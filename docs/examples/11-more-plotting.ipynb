# More plotting examples

```python
import climetlab as  cml
```

Get some GRIB data

```python
ds =  cml.load_dataset("sample-grib-data")
```

Plot the first field

```python
cml.plot_map(ds[0])
```

Plot the second field

```python
cml.plot_map(ds[1])
```

Plot both fields on the same map

```python
cml.plot_map((ds[0], ds[1]))
```

Alternative method:

```python
p = cml.new_plot()
p.plot_map(ds[0])
p.plot_map(ds[1])
p.show()
```

Switch off background map

```python
cml.plot_map(ds[1], background=False)
```

Switch off both foreground and background

```python
cml.plot_map(ds[1], foreground=False, background=False)
```

Projections

```python
cml.plot_map(None)
```

```python
cml.plot_map(None, projection="lambert")
```
