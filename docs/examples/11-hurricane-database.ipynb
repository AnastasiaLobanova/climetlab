# NOAA's hurricane database

```python
import climetlab as  cml
```

```python
atlantic = cml.load_dataset("hurricane-database", "atlantic")
```

```python
# cml.plotting_options(projection="mercator")
```

Get the Pandas frame

```python
df = atlantic.to_pandas()
```

## Huricanes making landfall at category 5

```python
df[(df.category==5) & (df.type=='L')]
```

## Get the track for Katrina

```python
katrina = df[(df.name=='katrina') & (df.year==2005)]
```

```python
katrina
```

## Plot the track

```python
cml.plot_map(katrina, margins=1.5)
```

## Retrieve matching fields from ERA5

```python
source = cml.load_source("cds", 
                         "reanalysis-era5-pressure-levels",
                         variable="z",
                         level=500,
                         product_type= "reanalysis",
                         time=[0,6,12,18],
                         based_on={"data": katrina[katrina.status == 'HU'], "extent": "bounding-box", "margins": 1.5})

for s in source:
    cml.plot_map(s, width=400, grid=True)
```

## Pacific bassin

```python
pacific = cml.load_dataset("hurricane-database", "pacific")
```

Get the Pandas frame

```python
df = pacific.to_pandas()
```

Plot a few years

```python
for year in range(2010, 2020):
    cml.plot_map(df[df.year == year], 
                 column='number',
                 title=year, 
                 margins=1, 
                 style='rainbow-markers')
```

## Plot each track independently

```python
C = ['red', 'green', 'blue',  'orange', 'purple', 'black']
p = cml.new_plot(margins=1)
i = 0
for year in range(2010, 2020):
    cp = df[(df.year == year) & (df.number == 1) & (df.bassin == 'CP')]
    if len(cp):
        p.plot_map(cp, style={'+symbol_colour': C[i]})
        i += 1
p.show()
```
