# Retrieve ERA5 data from the CDS

```python
import climetlab as  cml
```

```python
source = cml.load_source("cds", 
                         "reanalysis-era5-single-levels",
                         variable=["2t", "msl"],
                         product_type= "reanalysis",
                         area=[50, -50, 20, 50],
                         date= "2012-12-12",
                         time= "12:00")
for s in source:
    cml.plot_map(s)
```

```python
source = cml.load_source("cds", 
                         "reanalysis-era5-single-levels",
                         variable=["2t", "msl"],
                         product_type= "reanalysis",
                         area=[50, -50, 20, 50],
                         date= "2012-12-12",
                         time= "12:00",
                         format="netcdf")
for s in source:
    cml.plot_map(s)
```
