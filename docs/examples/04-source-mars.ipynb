# Retrieve data ECMWF MARS archive

```python
import climetlab as cml
```

```python
source = cml.load_source("mars", 
                         param=["2t", "msl"],
                         levtype="sfc",
                         area=[50, -50, 20, 50],
                         grid=[1, 1],
                         date="2012-12-13")
for s in source:
    cml.plot_map(s)
```
